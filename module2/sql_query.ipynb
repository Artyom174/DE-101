С помощью SQL-запросов вытянуть с таблицы следущую информацию:
* Total Sales - Общая сумма продаж
* Total Profit - Общая прибыль
* Profit Ratio - коэфицент прибыли
* Profit per Order - прибыль на заказ
* Sales per Customer - продажи на клиента
* Avg. Discount - средний % скидки
* Monthly Sales by Segment - месячные продажи по сегментам
* Monthly Sales by Product Category - месячные продажи по категории продукта
* Sales and Profit by Customer - Продажи и прибыль по клиентам
* Customer Ranking - ранг пользователей
* Sales per region - продажи на регион

```python
#общие продажи
query_total_sales = '''
select 
    round(sum(sales),2) as total_sales
from orders
'''
#общая прибыль
query_total_profit = '''
select 
    round(sum(profit),2)  as total_profit
from orders
'''
#коэфицент прибыли
query_profit_ratio = '''
select 
    round(sum(profit)/sum(sales),2) as profit_ratio
from orders
'''
#прибыль на заказ
query_profit_per_order = '''
select 
    round(sum(profit)/count(order_id),2) as profit_per_order
from orders
'''
#сумма продаж на покупателя
query_sales_per_customer = '''
select 
    round(sum(sales)/count(customer_id),2) as sales_per_customer
from orders
'''
#средний % скидки
query_avg_discount = '''
select 
    round(avg(discount),2) as avg_discount
from orders
'''
#месячные продажи по сегментам
query_monthly_sales_by_segment = '''
select 
    extract(month from order_date) as month,
    segment,
    sum(sales) as month_sales
from orders
group by 1,2
order by 1
'''
#месячные продажи по категории продукта
query_monthly_sales_by_category = '''
select 
    extract(month from order_date) as month,
    category,
    sum(sales) as month_sales
from orders
group by 1,2
order by 1
'''
#Продажи, прибыль и ранг по клиентам
query_sales_profit_by_customer ='''
select 
    customer_name,
    round(sum(sales), 2) as total_sales,
    round(sum(profit), 2) as total_profit
from orders
group by 1
order by 3 desc
'''
#продажи на регион
query_sales_per_region ='''
select
    region,
    round(sum(sales),2) as total_sales
from orders
group by 1
'''
```
